## Converting to gray scale Image

* R --> Red channel values
* G --> Green Channel values
* B --> Blue Channel values

```python
R, G, B = image[:, :, 0], image[:, :, 1], image[:, :, 2]
image_1 = (R+G+B) / 3

plt.imshow(image_1, cmap='gray');
```
